## ðŸ“ˆ **[Activation Function](https://www.tensorflow.org/api_docs/python/tf/keras/activations?msclkid=fc4a78a0d01111ecad731c12a9dedc5d) Overview**

  * ### Activation functions take the **output of a neuron** as its domain and **maps it to some known range**. 
  * ### Activation functions are **relevent durring 3 stages**:
    * ### Passing information **during the forward pass** for inference. 
    * ### Passing gradients for **backpropogation to update model weights** during training.
    * ### Providing ways to **encode the predictions on the output neuron**.
    * ### examples:
      * ### Inner Layers: **Linear, Relu, Tanh**
      * ### Output: **Softmax, Sigmoid**  

```python
from tensorflow import constant,GradientTape
from tensorflow.keras import activations
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
x = constant(linspace(-10,10,1000),dtype='float64')
```

```python
data = {'linear':activations.linear(x),
        'relu':activations.relu(x),
        'elu':activations.elu(x),
        'selu':activations.selu(x),
        'gelu':activations.gelu(x),
        'swish':activations.swish(x),
        'exponential':activations.exponential(x),
        'tanh':activations.tanh(x),
        'sigmoid':activations.sigmoid(x),
        'hard sigmoid':activations.hard_sigmoid(x),
        'softplus':activations.softplus(x),
        'softsign':activations.softsign(x)
        }
```

```python
afuncs = {'linear':activations.linear,
        'relu':activations.relu,
        'elu':activations.elu,
        'selu':activations.selu,
        'gelu':activations.gelu,
        'swish':activations.swish,
        'exponential':activations.exponential,
        'tanh':activations.tanh,
        'sigmoid':activations.sigmoid,
        'hard sigmoid':activations.hard_sigmoid,
        'softplus':activations.softplus,
        'softsign':activations.softsign
        }
data_y = {}
data_dy = {}

```

```python
#------- Generate data:
for name,func in afuncs.items():
  with GradientTape() as tape:
    tape.watch(x)
    y = func(x)
    data_y[name] = y
  data_dy[name] = tape.gradient(y,x)
```

## ðŸ”‘ Activation functions are relevant during forward pass & output. 

```python
fig, axes = plt.subplots(4,3,figsize=(11,13),sharex=True)
axes = axes.flatten()
for ax,items in zip(axes,data_y.items()):
  ax.plot(x,items[1],color='tab:blue',label='y')
  ax.set_title(items[0])
  ax.legend()

text = fig.suptitle('Activation Functions',fontsize=15,y=1.0)
fig.tight_layout(pad=1.5,h_pad=1.20,w_pad=1.0)
```

## ðŸ”‘ Derivatives of the activation functions are relevant during backpropogation.

```python
fig, axes = plt.subplots(4,3,figsize=(11,13),sharex=True)
axes = axes.flatten()
for ax,items in zip(axes,data_dy.items()):
  ax.plot(x,items[1],color='tab:red',label='y')
  ax.set_title(items[0])
  ax.legend()

text = fig.suptitle('Gradient Functions',fontsize=15,y=1.0)
fig.tight_layout(pad=1.5,h_pad=1.20,w_pad=1.0)
```
